## Deep Learning with Neural Networks using TensorFlow

In this first program Deep Learning with Neural Networks using TensorFlow, we will start with simple program to get familiarize with the libraries, basic building blocs and how the overall program works.

## Step 1 - Import Libraries

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
# Check the version of tensorflow
print(tf.__version__)
```

## Step 2 - Define Data
Define numpy array for dependent or target (y) feature and the indepedent (x) feature to be used in the TensorFlow model.

```python
x = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0])
y = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0])
```

## Step 3 - Build and Compile Neural Network Model
We are going to build a very simple keras sequential model with one neuron and 1 layer. Then we are going to compile the model with certain type of loss and optimizer, which helps the model keep iterating while minimizing loss till it gets the best model.

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

## Step 4 - Fit the model with data
Here the model will be fed with both dependent and indepedent features of the training data. The model will iterate 500 times as defined by the epochs parameter to find the best model.

```python
model.fit(x, y, epochs=500)
```

## Step 5 - Use the trained model on unknown test data to predict the result.

```python
x_test = [10.0]
y_pred = model.predict(x_test)
print(y_pred)
```
